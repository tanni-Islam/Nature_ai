```python
# function to generate a random number between two given numbers
def rand(a,b):
    return (b-a)* random.random() + a
```

```python
# Make matrix
def makeMatrix(a,b):
    return np.zeros([a,b])
```

```python
# out sigmoid function
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
# derivative of sigmoid
def dsigmoid(y):
    return y*(1-y)
```

```python
class NN:
    # our pocha-docha NN
    def __init__(self, no_i, no_h, no_o):
        #number of inputs, hidden and output nodes
        self.no_i = no_i + 1
        self.no_h = no_h
        self.no_o = no_o
        
        # activation for nodes
        self.act_i = [1.0] * self.no_i
        self.act_h = [1.0] * self.no_h
        self.act_o = [1.0] * self.no_o
        
        # create weight 
        
        self.weight_i = makeMatrix(self.no_i, self.no_h)
        self.weight_o = makeMatrix(self.no_h, self.no_o)
        
        #random weights for input and hidden layes
        for i in range(self.no_i):
            for j in range(self.no_h):
                self.weight_i[i][j] = rand(-0.2, 0.2)
        
        #random weights for hidden and output layer
        for j in range(self.no_h):
            for k in range(self.no_o):
                self.weight_o[j][k] = rand(-2.0, 2.0)
        
        self.change_i = makeMatrix(self.no_i, self.no_h)
        self.change_o = makeMatrix(self.no_h, self.no_o)
        
        pass
    
    def activate(self, inputs):
        
        #input activation
        for i in range(self.no_i - 1):
            self.act_i[i] = inputs[i]
        #hidden activation
        for j in range(self.no_h):
            i_sum=0.0
            for i in range(self.no_i):
                i_sum = i_sum + self.act_i[i]*self.weight_i[i][j] 
            self.act_h[j] = sigmoid(i_sum)
            
        #output activation
        for k in range(self.no_o):
            o_sum = 0.0
            for j in range(self.no_h):
                o_sum = o_sum + self.act_h[j]*self.weight_o[j][k]
                
            self.act_o[k] = sigmoid(o_sum)
            
        pass
    
    def backprop(self, targets, N ,M):
        # calculate error for output layer
        output_deltas = np.zeros(self.no_o)
        for k in range(self.no_o):
            error = targets[k] - self.act_o[k]
            ouput_deltas[k] = dsigmoid(self.act_o[k]) #* error
        # calculate error for hidden layer
        hidden_deltas = np.zeros(self.no_h)
        for j in range(self.no_h):
            error_h = 0.0
            for k in range(self.no_o):
                error_h = error_h + output_deltas[k] * self.weight_o[j][k]
            hidden_deltas[j] = dsigmoid(self.act_h[j]) * error_h
            
        #update output weights
        for j in range(self.no_h):
            for k in range (self.no_o):
                change = output_deltas[k] * self.act_h[j]
                self.weight_o[j][k] = self.weight_o[j][k] + N * change + M * self.change_o[j][k]
                self.change_o[j][k] = change
        
        #update input weights
        for i in range(self.no_i):
            for j in range(self.no_h):
                change = hidden_deltas[j] * self.act_i[i]
                self.weight_i[i][j]= self.weight_i[i][j] + N * change + M * self.change_i[i][j]
                self.change_i[i][j] = change
                
        #calculate ultimate loss using euclidian distance
        loss = 0.0
        for k in range(len(targets)):
            loss = loss + 0.5 * (targets[k] - self.act_o[k]) ** 2
        
        return loss
    
    def test(self, patterns):
        self.predict = np.empty([len(patterns), self.no_o])
        for i,p in enumerate(patterns):
            self.predict[i] = self.activate(p)
        pass
    def train(self, patterns, iterations = 1000, N = 0.5, M = 0.1):
        
        patterns = list(patterns)
        
        for i in range(iterations):
            loss = 0.0
            for p in patterns:
                inputs = p[0]
                targets = p[1]
                self.activate(inputs)
                loss = loss +  
        pass
    
```
